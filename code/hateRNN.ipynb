```python
comments = data['comments']
hate = data['hate']
print(len(data))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(comments,hate,stratify=hate, test_size= 0.2, random_state=12)
```

```python
print(len(X_train))
print(len(X_test))

y_train = np.asarray(y_train)
print(X_train)
```

```python
from konlpy.tag import Okt
t= Okt()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
token_train = []
token_test = []
```

```python
for i in X_train:
  token_train.append(t.morphs(i))
```

```python
for i in X_test:
  token_test.append(t.morphs(i))
```

```python
print(token_test)
```

```python
print(token_train)
print(token_test)
```

```python
#train encoding
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(token_train)
train_encoded = tokenizer.texts_to_sequences(token_train)
print(train_encoded)
```

```python
#test encoding
test_encoded = tokenizer.texts_to_sequences(token_test)
print(test_encoded)
```

```python
#train padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
train_padded = pad_sequences(train_encoded,maxlen= 50)
```

```python
#test padding
test_padded = pad_sequences(test_encoded,maxlen= 50)
```

```python
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.layers import SimpleRNN, Embedding, Dropout, LSTM, Bidirectional
from tensorflow.keras.models import Sequential

num_words = 18000


model= Sequential()
model.add(Embedding(num_words, 64))
model.add(Bidirectional(LSTM(64)))
model.add(Dense(1, activation='sigmoid')) #최종 결과확인

model.compile(optimizer ='rmsprop',
              loss='binary_crossentropy',
              metrics=['acc'])

model.summary()
```

```python
history= model.fit(train_padded, y_train,
                   epochs=15,
                   batch_size= 128,
                   validation_split=0.2)
```

```python
score= model.evaluate(test_padded, y_test)
print(f'#Test accuracy:{score[1]:.3f}')
```

```python
model.predict(test_padded)
```

```python
#문장 들어오면 토큰화 + encoding + padding 후 모델에 돌려보기

def padding(text) :
  morp = t.morphs(text)
  encoded = tokenizer.texts_to_sequences([morp])
  padded = pad_sequences(encoded,maxlen= 50)
  return padded
```
