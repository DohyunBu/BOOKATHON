```python
import pandas as pd

def ratio(x):
    num_re = 0
    num_to = x.count(" ")
    word_list = ['사랑', '여자', '엄마', '마음', '낭만', '순정', '사모', '애정', '연애']

    for i in word_list:
        num_re += x.count(i)
    return num_re/num_to

df1['ratio']=df1['lyrics'].apply(ratio) # csv 파일 내부의 글 내용('lyrics') -> 연관률('ratio') 라벨링
df1 = df1.sort_values(by='ratio',ascending=False) # 연관률 순서대로 나열
```
