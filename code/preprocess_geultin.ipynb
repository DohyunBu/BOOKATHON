```python
banned_title_list = ['장원','필독','공지','글틴']
banned_content_list=['저는 ', '제가 ', '습니다', '입니다']

banned_index = []
for i in range(len(data)):
    for word in banned_title_list:
        if word in str(data.iloc[i,1]):
            banned_index.append(i)
            # print(data.iloc[i,1])
            break
            
    for word in banned_content_list:
        if word in df.iloc[i,1]:
            banned_index.append(i)
            break

newframe= df.drop(index=banned_index, axis=0)
newframe= newframe.reset_index(drop=True)

newframe
```

```python
import re
from hanspell import spell_checker
li = ["출처","약력","발행일"]


for i in range(len(newframe)):
    
    k = newframe.loc[i,'content']
    
    # 정규표현식을 이용한 1차 전처리
    k = re.sub(r"(http|https)?:\/\/\S+\b|www\.(\w+\.)+\S*", '', k)# url주소 제거
    k = re.sub("[a-zA-Z0-9+-_.]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+$']", '',k)#이메일제거
    k = re.sub('[0-9]{4}년 [0-9]{,2}월 [0-9]{,2}일','',k) #날짜 제거
    k = re.sub('[0-9]{4}년 [0-9]{,2}월','',k) #날짜 제거
    k = re.sub('[0-9]{4}. [0-9]{,2}. [0-9]{,2}','',k) #날짜 제거
    k = re.sub(r"\([^)]*\)",'',k) #() 안, () 제거
    k = re.sub(r"\[[^)]*\]",'',k) #[] 안, [] 제거
    k = re.sub(r"\<[^)]*\>",'',k) #[] 안, [] 제거
    k = re.sub('[ぁ-ゔァ-ヴー々〆〤一-龥]','',k) #한자, 일본어 제거 
    k = re.sub('[ㄱ-ㅎㅏ-ㅣ]+','',k) #자음, 모음 제거
    k = re.sub('“','"',k) #큰 따옴표 통일
    k = re.sub('”','"',k)
    k = re.sub('‘',"'",k) #작은 따옴표 통일
    k = re.sub('’',"'",k)
    k = re.sub('[^가-힣 0-9 a-z A-Z \. \, \' \" \? \! ]+','',k) #특수문자 제거
    
    newframe.loc[i,'content'] = k
 
    
    new=""

    k=k.replace(".",".#")
    k=k.replace("?","?#")
    k=k.replace("!","!#")
    
    sen_list = [x.strip() for x in k.split("#")]
    for sen in sen_list:
        if sen.startswith("이 글은") and sen.endswith("밝혀 둡니다."):
    k= spell_checker.check(k).as_dict()['checked']
        a= any(keyword in new_sen for keyword in li) #특정 키워드(리스트 li에) 찾아서 제거
        if a is True:
            continue
        else:
            new = new +" " + new_sen
    print(new)
    newframe.loc[i,'content'] = k

newframe
```
